```python
import warnings
import sys
warnings.filterwarnings('ignore')

import numpy as np, pandas as pd
from numpy import inf

sys.setrecursionlimit(5000)
```

```python
Oi = np.array([100, 120, 140])
Dj = np.array([80, 150, 130])
W_data = np.array([[0, 7, 2],[7, 0, 6],[2, 6, 0]])

Oi = np.round(pd.read_excel('./data/supply.xlsx').T.values[0], decimals=4)
Dj = pd.read_excel('./data/demand.xlsx').T.values[0]
W_data = pd.read_excel('./data/weight.xlsx').to_numpy()

Oi
```

```python
# Create first temp data
W_temp = (Dj * Oi[np.newaxis].T) / W_data
W_temp[W_temp == inf] = 0
Ai = Oi / W_temp.sum(axis=1)
W_temp = W_temp * Ai[np.newaxis].T

print(type(W_temp))
```

```python
# Itteration function
def iteration(temp_data, filepath, type = 0, num = 1):
    print("Itteration {}".format(num))
    Dd = temp_data.sum(axis=type)
    if type == 0:
        if (Dd==Dj).all() == False:
            Bj = Dj / Dd
            temp_data = temp_data * Bj
            num = num + 1
            iteration(temp_data, filepath, 1, num)
        else:
            print(temp_data)
            df = pd.DataFrame(temp_data)
            df.to_excel(filepath, index=False)
    else:
        if (Dd==Oi).all() == False:
            Bj = Oi / Dd
            temp_data = temp_data * Bj[np.newaxis].T
            num = num + 1
            iteration(temp_data, filepath, 0, num)
        else:
            print(temp_data)
            df = pd.DataFrame(temp_data)
            df.to_excel(filepath, index=False)
```

```python
iteration(W_temp, "./data/result_15.xlsx")
```
